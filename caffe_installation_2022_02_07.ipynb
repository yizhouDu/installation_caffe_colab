# 5. Make necessary Adjustments

```python
# replace the paths with your own
!export PYTHONPATH=/content/drive/MyDrive/temp/caffe/python:$PYTHONPATH
!export PYTHONPATH=/content/drive/MyDrive/temp/caffe:$PYTHONPATH
!chmod 755 -R /content/drive/MyDrive/temp
!chmod 755 -R /content/drive/MyDrive/temp/caffe
```

```python
!sed -i 's/-gencode arch=compute_20/#-gencode arch=compute_20/' Makefile.config # old cuda versions won't compile 
!sed -i 's/\/usr\/local\/include/\/usr\/local\/include \/usr\/include\/hdf5\/serial\//'  Makefile.config #one of the 4 things needed to fix hdf5 issues
!sed -i 's/# OPENCV_VERSION := 3/OPENCV_VERSION := 3/' Makefile.config # We actually use opencv 4.1.2, but it's similar enough to opencv 3.
!sed -i 's/code=compute_61/code=compute_61 -gencode=arch=compute_70,code=sm_70 -gencode=arch=compute_75,code=sm_75 -gencode=arch=compute_75,code=compute_75/' Makefile.config # support for new GPUs
!sed -i 's/# PYTHON_LIBRARIES := boost_python3 python3.5m/PYTHON_LIBRARIES := boost_python-py36/' Makefile.config # Adjust boost version to the current one
!sed -i 's/PYTHON_INCLUDE := \/usr\/include\/python2.7 \\/PYTHON_INCLUDE := \/usr\/include\/python3.7 \\/' Makefile.config # Adjust path to find python.h and numpy/arrayobject.h
!sed -i 's/		\/usr\/lib\/python2.7\/dist-packages\/numpy\/core\/include/		\/usr\/local\/lib\/python3.7\/dist-packages\/numpy\/core\/include/' Makefile.config
```
